```python
!pip -q install pysus pandas pyarrow

from pathlib import Path
from pysus.online_data.SIH import SIH

OUTDIR = Path("/content/datasus_pysus_sih_rd")
OUTDIR.mkdir(parents=True, exist_ok=True)

UF_LIST = ["AC","AL","AP","AM","BA","CE","DF","ES","GO","MA","MT","MS","MG","PA","PB",
           "PR","PE","PI","RJ","RN","RS","RO","RR","SC","SP","SE","TO"]

def baixar_sih_rd_uf(uf: str, year: int, months=range(1,13)):
    sih = SIH().load()
    uf_dir = OUTDIR / uf
    uf_dir.mkdir(parents=True, exist_ok=True)

    for m in months:
        out_csv = uf_dir / f"SIH_RD_{uf}_{year}_{m:02d}.csv"
        if out_csv.exists():
            print("Já existe, pulando:", out_csv.name)
            continue

        files = sih.get_files("RD", uf=uf, year=year, month=[m])
        if not files:
            print(f"[{uf}] sem arquivo {year}-{m:02d}")
            continue

        print(f"[{uf}] baixando {year}-{m:02d} ->", files)
        pqset = sih.download(files, local_dir=str(uf_dir))
        df = pqset.to_dataframe()
        df.to_csv(out_csv, index=False)
        print(f"[{uf}] salvo:", out_csv.name, "shape=", df.shape)

# =========================
# 1) Primeiro só MG (2018 inteiro)
# =========================
baixar_sih_rd_uf("MG", 2018)

# =========================
# 2) Depois Brasil todo (2018 inteiro)
# =========================
for uf in UF_LIST:
    if uf == "MG":
        continue
    baixar_sih_rd_uf(uf, 2018)
```
